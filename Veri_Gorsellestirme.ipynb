```python
#--------ÇİZGİ GRAFİĞİ (LİNE CHART)--------

import pandas as pd  #Pandas: Verileri tablolar halinde işlemek için 
import matplotlib.pyplot as plt  #Matplotlib: Grafik çizmek için kullandığımız kütüphane  

# 1. Önce örnek bir veri oluşturalım 
data = {'yas': [20, 18, 32, 35, 33, 48, 38, 38, 62, 55]}
veriSeti = pd.DataFrame(data)

# x ekseni 
x = range(1, 11)
# 1'den 10'a kadar sayıları oluşturur. Grafik noktalarının sayısını belirler.

# y ekseni 
y = veriSeti.iloc[0:10, 0]
# iloc (index location) şunu der: "Tablonun ilk on satırını (0:10) ve 1. sütunun (0) al."

# Grafik çizimi 
plt.plot(x, y, "o:r")
# "o" -> Noktalar yuvarlak olsun.
# ":" -> Çizgi kesikli (noktalı) olsun.
# "r" -> Renk kırmızı (red) olsun.
plt.show() 
# oluşturulan grafiği ekranda gösterir.
```

```python
#--------ÇİZGİ GRAFİĞİ (LİNE CHART)--------

import numpy as np # Matematiksel işlemler ve düzenli sayı dizileri(x1gibi) oluşturmak için
import pandas as pd 
import matplotlib.pyplot as plt

data = {'yas': [20, 18, 32, 35, 33, 48, 38, 38, 62, 55, 
                22, 25, 30, 28, 40, 45, 50, 55, 60, 65,
                19, 21, 23, 25, 27, 29, 31, 33, 35, 37,
                40, 42, 44, 46, 48, 50, 52, 54, 56, 58]}
veriSeti = pd.DataFrame(data)

x1 = np.arange(1, 21)        # 20 adet sayı üretir

y1 = veriSeti.iloc[0:20, 0]  # Tablodan 20 adet veri çeker (0-20 arası)
y2 = veriSeti.iloc[20:40, 0] # Tablodan 20 adet veri çeker (20-40 arası)

plt.xticks(x1)  # Alt eksendeki sayıları  1,2,3....20 diye sabitler               
plt.plot(x1, y1, x1, y2)  # İki farklı veri grubunu aynı grafiğe çizer.
plt.show()
```

```python
#--------SÜTUN GRAFİĞİ (BAR CHART)--------


import pandas as pd
import matplotlib.pyplot as plt 


data = {
    'sigaraDurum': ['hayir', 'evet', 'hayir', 'evet'],
    'odemeMiktari': [13270, 32000, 11000, 35000]
}
veriSeti = pd.DataFrame(data)


# sigara içenleri ve içmeyenlerin ödeme ortalamasını hesaplar
ozet = veriSeti.groupby("sigaraDurum")["odemeMiktari"].mean()

# x: Kategoriler (evet/hayır), height: Çubukların uzunluğu 
plt.bar(x=ozet.index, height=ozet.values, color='r')


plt.xlabel("Sigara İçme  Durumu")
plt.ylabel("Ödeme Miktarı")
plt.title("Sigara Durumuna Göre Ortalama Ödeme")

plt.show()
```

```python
#--------PASTA GRAFİĞİ (PİE CHART)--------


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


#--------VERİ SETİ HAZIRLAMA--------
data = {
    'bolge': ['kuzeybati', 'kuzeydogu', 'guneybati', 'guneydogu', 'guneydogu', 'kuzeybati'],
    'odemeMiktari': [1000, 1200, 800, 1500, 1000, 900]
}
veriSeti = pd.DataFrame(data)


#--------VERİYİ GRUPLAMA--------
# 'bolge' sütununa göre grupla ve her bölgenin  'oodemeMiktarı' toplamını al
ozet = veriSeti.groupby("bolge")["odemeMiktari"].sum()


#--------GÖRSEL AYARLAMA--------
etiketler = ozet.index  # Bölgelerin isimleri (kuzeybatı vb.)
degerler = ozet.values  # Hesaplana toplam para miktarları
renkler = sns.color_palette("viridis", len(etiketler)) # profesyonel renk paleti
secim = (0, 0, 0, 0.2)  # 4. dilimi (Güneydoğu) %20 oranında dışarı çıkar (vurgula)


#--------PASTA GRAFİĞİ ÇİZME-------- 
plt.figure(figsize=(8, 7)) # grafiğin boyutunu ayarlar
plt.pie(
    degerler,
    explode=secim,  # Vurgulanacak dilim
    labels=etiketler, # Dilim isimleri
    autopct="%1.1f%%", # Yüzde oranlarını yaz
    shadow=True, # garfiğin başlanngıç açısını döndürür
    startangle=140,
    colors=renkler # renkleri uygula
)


#--------TAMAMLAMA--------

plt.title("Bölgelere Göre Toplam Sigorta Ödemesi"), # başlığı ekle
plt.show() # grafiği ekrana bas
```
