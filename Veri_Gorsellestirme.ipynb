```python
#--------ÇİZGİ GRAFİĞİ (LİNE CHART)--------

import pandas as pd  #Pandas: Verileri tablolar halinde işlemek için 
import matplotlib.pyplot as plt  #Matplotlib: Grafik çizmek için kullandığımız kütüphane  

# 1. Önce örnek bir veri oluşturalım 
data = {'yas': [20, 18, 32, 35, 33, 48, 38, 38, 62, 55]}
veriSeti = pd.DataFrame(data)

# x ekseni 
x = range(1, 11)
# 1'den 10'a kadar sayıları oluşturur. Grafik noktalarının sayısını belirler.

# y ekseni 
y = veriSeti.iloc[0:10, 0]
# iloc (index location) şunu der: "Tablonun ilk on satırını (0:10) ve 1. sütunun (0) al."

# Grafik çizimi 
plt.plot(x, y, "o:r")
# "o" -> Noktalar yuvarlak olsun.
# ":" -> Çizgi kesikli (noktalı) olsun.
# "r" -> Renk kırmızı (red) olsun.
plt.show() 
# oluşturulan grafiği ekranda gösterir.
```

```python
#--------ÇİZGİ GRAFİĞİ (LİNE CHART)--------

import numpy as np # Matematiksel işlemler ve düzenli sayı dizileri(x1gibi) oluşturmak için
import pandas as pd 
import matplotlib.pyplot as plt

data = {'yas': [20, 18, 32, 35, 33, 48, 38, 38, 62, 55, 
                22, 25, 30, 28, 40, 45, 50, 55, 60, 65,
                19, 21, 23, 25, 27, 29, 31, 33, 35, 37,
                40, 42, 44, 46, 48, 50, 52, 54, 56, 58]}
veriSeti = pd.DataFrame(data)

x1 = np.arange(1, 21)        # 20 adet sayı üretir

y1 = veriSeti.iloc[0:20, 0]  # Tablodan 20 adet veri çeker (0-20 arası)
y2 = veriSeti.iloc[20:40, 0] # Tablodan 20 adet veri çeker (20-40 arası)

plt.xticks(x1)  # Alt eksendeki sayıları  1,2,3....20 diye sabitler               
plt.plot(x1, y1, x1, y2)  # İki farklı veri grubunu aynı grafiğe çizer.
plt.show()
```

```python
#--------SÜTUN GRAFİĞİ (BAR CHART)--------


import pandas as pd
import matplotlib.pyplot as plt 


data = {
    'sigaraDurum': ['hayir', 'evet', 'hayir', 'evet'],
    'odemeMiktari': [13270, 32000, 11000, 35000]
}
veriSeti = pd.DataFrame(data)


# sigara içenleri ve içmeyenlerin ödeme ortalamasını hesaplar
ozet = veriSeti.groupby("sigaraDurum")["odemeMiktari"].mean()

# x: Kategoriler (evet/hayır), height: Çubukların uzunluğu 
plt.bar(x=ozet.index, height=ozet.values, color='r')


plt.xlabel("Sigara İçme  Durumu")
plt.ylabel("Ödeme Miktarı")
plt.title("Sigara Durumuna Göre Ortalama Ödeme")

plt.show()
```

```python
#--------PASTA GRAFİĞİ (PİE CHART)--------


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


#--------VERİ SETİ HAZIRLAMA--------
data = {
    'bolge': ['kuzeybati', 'kuzeydogu', 'guneybati', 'guneydogu', 'guneydogu', 'kuzeybati'],
    'odemeMiktari': [1000, 1200, 800, 1500, 1000, 900]
}
veriSeti = pd.DataFrame(data)


#--------VERİYİ GRUPLAMA--------
# 'bolge' sütununa göre grupla ve her bölgenin  'oodemeMiktarı' toplamını al
ozet = veriSeti.groupby("bolge")["odemeMiktari"].sum()


#--------GÖRSEL AYARLAMA--------
etiketler = ozet.index  # Bölgelerin isimleri (kuzeybatı vb.)
degerler = ozet.values  # Hesaplana toplam para miktarları
renkler = sns.color_palette("viridis", len(etiketler)) # profesyonel renk paleti
secim = (0, 0, 0, 0.2)  # 4. dilimi (Güneydoğu) %20 oranında dışarı çıkar (vurgula)


#--------PASTA GRAFİĞİ ÇİZME-------- 
plt.figure(figsize=(8, 7)) # grafiğin boyutunu ayarlar
plt.pie(
    degerler,
    explode=secim,  # Vurgulanacak dilim
    labels=etiketler, # Dilim isimleri
    autopct="%1.1f%%", # Yüzde oranlarını yaz
    shadow=True, # garfiğin başlanngıç açısını döndürür
    startangle=140,
    colors=renkler # renkleri uygula
)


#--------TAMAMLAMA--------

plt.title("Bölgelere Göre Toplam Sigorta Ödemesi"), # başlığı ekle
plt.show() # grafiği ekrana bas
```

```python
#--------SERPİLME DİYAGRAMI--------


import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns # Gelişmiş grafik kütüphanesi (renklendirme ve estetik için)


#--------VERİ SETİNİ HAZIRLAMA--------
# 1000 kişilik veri oluşturuyoruz.
np.random.seed(42) # Her çalıştırıldığında aynı grafiği görmeni sağlar.
N = 1000

# Rastegele Yaşlar(18-65)
yas = np.random.randint(18, 65, N)


# Rastgele Sigara Durumu (%20 evet, %80 Hayır)
sigaraDurum = np.random.choice(['evet', 'hayir'], N, p=[0.2, 0.8])


# Temel Ödeme (Yaşa bağlı artış)
# Her yaş için 250 birim artış + 5000 birim taban fiyat.
temel_odeme = 5000 + (yas * 250)


# Sigara İçenler İçin Ekstra Maliyet
# Eğer sigara 'evet' ise, ödemeye 15.000 ile 35.000 arasında rastgele büyük bir rakam ekle.
# Eğer sigara 'hayir' ise, sadece 0 ile 5000 arası küçük bir gürültü ekle.
ekstra_maliyet = np.where(sigaraDurum == 'evet',
                   np.random.uniform(15000, 35000, N), # Sigara içenler uçuşuyor
                   np.random.uniform(0, 5000, N) # Sigara içmeyenler şerit halinde
                          )

# toplam ödeme 
odemeMiktari = temel_odeme + ekstra_maliyet


#--------VERİYİ DATAFRAME'E (tabloya) ÇEVİRİYORUZ--------
data = {
    'yas': yas,
    'sigaraDurum': sigaraDurum,
    'odemeMiktari': odemeMiktari
}
veriSeti = pd.DataFrame(data)


#--------SEABORN İLE GÖRSELLEŞTİRME--------
plt.figure(figsize=(12,8)) # Grafiğin boyutunu büyütüyoruz.(Çakışmaları engellemek için)


#seaborn serpilme diyagramı komutu
sns.scatterplot(
    x = "yas",
    y = "odemeMiktari",
    hue = "sigaraDurum", # Renklendirme: Sigara Durumu (evet/hayir)
    palette={'evet': 'orange', 'hayir': 'blue'},
    data=veriSeti #kullanılacak tablo
)


#--------ETİKETLEME--------
plt.xlabel("Yaş (yas)")
plt.ylabel("Ödeme Miktari (odemeMiktari)")
plt.title("Sigara Durumuna Göre Yaş ve Ödeme İlişkisi")
plt.legend(title="Sigara İçme Durumu (sigaraDurumu)") # Sağ üstteki kutu başlığı

plt.show()
```

```python
#--------HİSTOGRAM--------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#--------VERİ SETİNİ HAZIRLAMA--------
#VKI genellikle "Normal Dağılım"(Çan Eğrisi) gösterir.
np.random.seed(42)
vki_verisi = np.random.normal(loc=30, scale=6, size=1338) # Ortalama 30, standart sapma 6 olan 1338 veri

veriSeti = pd.DataFrame({'vki': vki_verisi})

#--------PEMBE HİSTOGRAM (DİKEY)--------
plt.figure(figsize=(8, 5))
sns.histplot(
    data=veriSeti,
    x="vki",     # x eksenine vki değerlerini koyar.
    color="hotpink"  # Grafiğin rengini pembe yapar
)
plt.title("VKI Dağılımı (Dikey)")
plt.xlabel("Vücut Kitle İndeksi (vki)")
plt.ylabel("Kişi Sayısı (Count)")
plt.show

```

```python
#--------BOX PLOT (KUTU GRAFİĞİ)--------
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
# Boyut hatası almamak için labels=[...] şeklinde liste kullanıyoruz
bp = plt.boxplot(veriSeti["vki"], 
                 vert=True, 
                 showmeans=True, 
                 meanline=True, 
                 labels=['vki'], # Köşeli paranteze dikkat!
                 patch_artist=True,
                 medianprops={'linewidth': 2, 'color': 'purple'},
                 meanprops={'linewidth': 2, 'color': 'red'},
                 notch=True)

plt.title("3.6 Kutu Grafiği: VKI Özet İstatistikleri")
plt.ylabel("Değer Aralığı")
plt.legend([bp['medians'][0], bp['means'][0]], ['Medyan (Mor)', 'Ortalama (Kırmızı)'])
plt.show()
```

```python
# --------VIOLIN GRAFİĞİ--------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --------VERİ SETİ--------
np.random.seed(42) # Rastgeleliği sabitleyerek grafiğin hep aynı çıkmasını sağlar.
n_samples = 1000   # Tablodaki toplam kişi sayısı.

data = {
    'yas': np.random.randint(18, 65, n_samples),
    'sigaraDurumu': np.random.choice(['evet', 'hayir'], n_samples, p=[0.2, 0.8]),
    'vki': np.random.normal(30, 6, n_samples)
}
veriSeti = pd.DataFrame(data)


veriSeti['odemeMiktari'] = 5000 + (veriSeti['yas'] * 200) + \
    np.where(veriSeti['sigaraDurumu'] == 'evet', np.random.uniform(15000, 35000, n_samples), 0)


plt.figure(figsize=(10, 6))

sns.violinplot(
    y="sigaraDurumu",   # NOT: Kategorik veri (evet/hayir) dikey eksende.
    x="odemeMiktari",   # NOT: Sayısal veri (para) yatay eksende.
    data=veriSeti,      # NOT: Grafiğin çizileceği kaynak tablo.
    palette="coolwarm"  # NOT: Renklerin mavi-turuncu (soğuk-sıcak) olmasını sağlar.
)

plt.title("3.7 Violin Grafiği: Sigara Durumuna Göre Ödeme")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#-------VERİ SETİ TANIMLAMA--------
np.random.seed(42)
# Notta geçen tüm sütunları (yas, vki, cocukSayisi, odemeMiktari) oluşturuyoruz
data = {
    'yas': np.random.randint(18, 65, 100),
    'vki': np.random.normal(30, 5, 100),
    'cocukSayisi': np.random.randint(0, 5, 100),
    'odemeMiktari': np.random.uniform(5000, 50000, 100)
}
veriSeti = pd.DataFrame(data)

# --- ISI HARİTASI KODU ---
# 1. Önce değişkenler arasındaki ilişki katsayılarını hesaplar
korelasyon = veriSeti[["yas", "vki", "cocukSayisi", "odemeMiktari"]].corr()

plt.figure(figsize=(8, 6))

# 2. Isı haritasını çizer
sns.heatmap(
    korelasyon, 
    annot = True,    # Karelerin içine korelasyon rakamlarını yazar.
    square = True,   # Tabloyu tam kare formuna sokar.
    cmap = "Reds"    # Renk paletini kırmızı tonları yapar.
)

plt.show()
```
